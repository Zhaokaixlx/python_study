# LLE局部线性嵌入降维算法

```python
# manifold 流行
from sklearn.manifold import LocallyLinearEmbedding
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

## 创建数据（瑞士卷）

```python
X,y = datasets.make_swiss_roll(n_samples=1500,noise=0.05,random_state=248)

fig = plt.figure(figsize=(12,9))
axes3 = fig.add_subplot(projection = '3d')

axes3.scatter(X[:,0],X[:,1],X[:,2],c = y)
axes3.view_init(7,-80)
```

## PCA降维

```python
pca = PCA(n_components=0.95)
pca.fit(X)
X_pca = pca.transform(X)
display(X_pca.shape)
# 2维的图形
plt.scatter(X_pca[:,0],X_pca[:,1], c = y)
```

## LLE局部线性降维

```python
lle = LocallyLinearEmbedding(n_neighbors=10,n_components=2)
lle.fit(X)

X_lle = lle.transform(X)
display(X_lle.shape)
plt.scatter(X_lle[:,0],X_lle[:,1],c = y)
```

# GMM高斯混合模型

```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
# 高斯混合模型（使用EM算法迭代优化）
from sklearn.mixture import GaussianMixture
```

## 创建数据

```python
from matplotlib.font_manager import FontManager
fm = FontManager()
[font.name for font in fm.ttflist]
```

```python
X,y = datasets.make_blobs(n_samples=1500,
                    cluster_std=[1.0,2.5,0.5],random_state = 170)
plt.figure(figsize=(6,6))
plt.rcParams['font.family'] = 'Songti SC'
plt.rcParams['font.size'] =20
plt.scatter(X[:,0],X[:,1],c = y)
plt.title('原始数据',pad = 20)
```

 ## Kmeans聚类

```python
kmeans = KMeans(3)
kmeans.fit(X)
y_ = kmeans.predict(X)
plt.scatter(X[:,0],X[:,1],c = y_)
plt.title('Kmeans聚类效果',pad = 20)
```

## GMM

```python
gmm = GaussianMixture(n_components=3)
y_ = gmm.fit_predict(X)
plt.figure(figsize=(6,6))
plt.scatter(X[:,0],X[:,1],c = y_)
plt.title('GMM聚类效果',pad = 20)
```

## 画在一起

```python
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.rcParams['font.family'] = 'Songti SC'
plt.rcParams['font.size'] =20
plt.scatter(X[:,0],X[:,1],c = y)
plt.title('原始数据',pad = 20)

plt.subplot(1,3,2)
kmeans = KMeans(3)
kmeans.fit(X)
y_ = kmeans.predict(X)
plt.scatter(X[:,0],X[:,1],c = y_)
plt.title('Kmeans聚类效果',pad = 20)

plt.subplot(1,3,3)
gmm = GaussianMixture(n_components=3)
y_ = gmm.fit_predict(X)
plt.scatter(X[:,0],X[:,1],c = y_)
plt.title('GMM聚类效果',pad = 20)
```
